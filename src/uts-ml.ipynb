```python
import pandas as pd

# Membuat DataFrame dengan data mahasiswa dan detail biaya per kedatangan
data = {
    'Nama': ['Ani', 'Budi', 'Joni', 'Jono', 'Lono'],
    'Biaya_per_kedatangan': [30000, 35000, 20000, 25000, 15000],
    'Senin': [1, 0, 1, 1, 0],
    'Selasa': [1, 1, 0, 1, 1],
    'Rabu': [1, 1, 1, 0, 1],
    'Kamis': [0, 1, 1, 1, 0],
    'Jumat': [1, 1, 0, 1, 1],
    'Sabtu': [0, 0, 1, 0, 1],
    'Minggu': [1, 0, 1, 0, 1]
}

df = pd.DataFrame(data)

# Menghitung jumlah kedatangan tiap mahasiswa
df['Total_Kedatangan'] = df[['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu']].sum(axis=1)

# Menghitung biaya total tiap hari
days = ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat', 'Sabtu', 'Minggu']
daily_cost = {day: df[day] @ df['Biaya_per_kedatangan'] for day in days}

# a) Rata-rata mahasiswa datang pada minggu ini
avg_attendance = df['Total_Kedatangan'].mean()

# b) Hari dengan biaya tertinggi
highest_cost_day = max(daily_cost, key=daily_cost.get)

# c) Hari dengan biaya lebih dari 110000
days_above_110000 = [day for day, cost in daily_cost.items() if cost > 110000]

# d) Mahasiswa yang paling banyak datang ke kampus
most_frequent_attendee = df.loc[df['Total_Kedatangan'].idxmax(), 'Nama']

# e) Mahasiswa yang datang pada hari minggu
sunday_attendees = df.loc[df['Minggu'] == 1, 'Nama'].tolist()

# f) Biaya tertinggi dan terendah
highest_cost = df['Biaya_per_kedatangan'].max()
lowest_cost = df['Biaya_per_kedatangan'].min()

# g) Frekuensi datang tertinggi dan terendah
highest_frequency = df['Total_Kedatangan'].max()
lowest_frequency = df['Total_Kedatangan'].min()

# Menampilkan hasil
print("a) Rata-rata mahasiswa datang pada minggu ini:", avg_attendance)
print("b) Hari dengan biaya tertinggi:", highest_cost_day)
print("c) Hari dengan biaya lebih dari 110000:", days_above_110000)
print("d) Mahasiswa yang paling banyak datang ke kampus:", most_frequent_attendee)
print("e) Mahasiswa yang datang pada hari minggu:", sunday_attendees)
print("f) Biaya tertinggi:", highest_cost, "dan terendah:", lowest_cost)
print("g) Frekuensi datang tertinggi:", highest_frequency, "dan terendah:", lowest_frequency)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Membuat DataFrame dengan informasi fakultas, jumlah mahasiswa, dan akreditasi
fakultas = ["Bisnis", "D3 Perhotelan", "ICT", "Ilmu Komunikasi", "Seni dan Desain"]
jumlah_mahasiswa = [260, 28, 284, 465, 735]
akreditasi = ["A", "A", "B", "A", "A"]

info_mahasiswa = pd.DataFrame({
    'Fakultas': fakultas,
    'Jumlah_Mahasiswa': jumlah_mahasiswa,
    'Akreditasi': akreditasi
})

# Menampilkan DataFrame
print(info_mahasiswa)

# Membuat plot dengan matplotlib
plt.figure(figsize=(10, 6))
bars = plt.bar(info_mahasiswa['Fakultas'], info_mahasiswa['Jumlah_Mahasiswa'], color=['blue', 'green', 'red', 'cyan', 'magenta'])

# Menambahkan judul dan label sumbu
plt.title('Jumlah Mahasiswa per Fakultas')
plt.xlabel('Fakultas')
plt.ylabel('Jumlah Mahasiswa')

# Menambahkan nilai di atas setiap batang
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom') # va: vertical alignment

# Menampilkan plot
plt.show()
```
